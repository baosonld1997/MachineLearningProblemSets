```python
df["launched"] = df.apply(lambda row: row["launched"].split(' ')[0], axis = 1)
```

```python
df["launchedYear"] = df.apply(lambda row: int(row["launched"].split('-')[0]), axis = 1)
df["deadlineYear"] = df.apply(lambda row: int(row["deadline"].split('-')[0]), axis = 1)
```
